```python
#write for user defined function
#1-usa ,2-Europ , 3-asia
def origin(num):
  if num==1:
    return 'USA'
  elif num==2:
    return 'Europ'
  else:
    return 'Asia'
```

```python
dataset['origin']=dataset['origin'].apply(origin)
```
